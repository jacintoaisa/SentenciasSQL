1.- Queremos saber cuantos productos hay para cada categoria, el total de productos debe de ser de 504.

```sql
USE AdventureWorks2016
SELECT Production.ProductSubcategory.Name, COUNT(*) AS Cantidad
FROM     Production.ProductSubcategory RIGHT OUTER JOIN
                  Production.Product ON Production.ProductSubcategory.ProductSubcategoryID = Production.Product.ProductSubcategoryID
GROUP BY Production.ProductSubcategory.Name


USE AdventureWorks2016
SELECT distinct Production.ProductSubcategory.Name, Color, COUNT(*) over (PArtition by Production.ProductSubcategory.ProductSubcategoryID,Color ) AS Cantidad
FROM     Production.ProductSubcategory inner JOIN
                  Production.Product ON Production.ProductSubcategory.ProductSubcategoryID = Production.Product.ProductSubcategoryID


```

2.- Queremos saber que cantidad de un producto se ha vendido en total (El campo quantity), 

para cada producto, el total de productos debe de ser de 504

```sql
USE [AdventureWorks2016]

SELECT Production.Product.Name, COALESCE(SUM(Sales.SalesOrderDetail.OrderQty),0) AS Cantidad

FROM     Sales.SalesOrderDetail RIGHT OUTER JOIN

                  Production.Product ON Sales.SalesOrderDetail.ProductID = Production.Product.ProductID

GROUP BY Production.Product.Name

USE [AdventureWorks2016]

SELECT Production.Product.Name, SUM(Sales.SalesOrderDetail.OrderQty) OVER (PARTITION BY Production.Product.Name) AS Cantidad

FROM     Sales.SalesOrderDetail RIGHT OUTER JOIN

                  Production.Product ON Sales.SalesOrderDetail.ProductID = Production.Product.ProductID


```

\--3.- Queremos saber cual es el precio medio al que se ha vendido cada producto, 

\--la función avg sobre el campo Price.

```sql
SELECT Production.Product.ProductID, Production.Product.Name, AVG(Sales.SalesOrderDetail.UnitPrice) as Media

FROM     Sales.SalesOrderDetail INNER JOIN

                  Production.Product ON Sales.SalesOrderDetail.ProductID = Production.Product.ProductID

GROUP BY Production.Product.ProductID, Production.Product.Name

SELECT distinct Production.Product.ProductID, Production.Product.Name, AVG(Sales.SalesOrderDetail.UnitPrice) OVER (PARTITION by Production.Product.ProductID, Production.Product.Name) as Media

FROM     Sales.SalesOrderDetail INNER JOIN

                  Production.Product ON Sales.SalesOrderDetail.ProductID = Production.Product.ProductID

```

\--4.- Queremos saber cuales son los 10 productos más vendidos y los que menos, 

\--excluir en los que menos aquellos que no se han vendido.

```sql
SELECT  TOP 10 Production.Product.ProductID, Production.Product.Name, SUM(Sales.SalesOrderDetail.OrderQty) as Cantidad

FROM     Sales.SalesOrderDetail INNER JOIN

                  Production.Product ON Sales.SalesOrderDetail.ProductID = Production.Product.ProductID

GROUP BY Production.Product.ProductID, Production.Product.Name

ORDER BY Cantidad DESC

SELECT  TOP 10 Production.Product.ProductID, Production.Product.Name, SUM(Sales.SalesOrderDetail.OrderQty) as Cantidad

FROM     Sales.SalesOrderDetail INNER JOIN

                  Production.Product ON Sales.SalesOrderDetail.ProductID = Production.Product.ProductID

GROUP BY Production.Product.ProductID, Production.Product.Name

ORDER BY Cantidad ASC
```

\--5.- Los productos pertenecen a una clase, campo class de product quiero saber cuanto 

\--se ha vendido de cada clase

```sql
SELECT SUM(Sales.SalesOrderDetail.OrderQty) AS Cantidad, Production.Product.Class

FROM     Production.Product INNER JOIN

                  Sales.SalesOrderDetail ON Production.Product.ProductID = Sales.SalesOrderDetail.ProductID

GROUP BY Production.Product.Class
```

\--14.- Quiero saber cuanto se ha vendido de cada color (teniendo en cuenta transparentes) y que me salgan sólo

\--los tres colores que más se han vendido.

```sql
SELECT TOP 3 SUM(Sales.SalesOrderDetail.OrderQty) AS Cantidad, COALESCE(Production.Product.Color,'Transparente') as Color

FROM     Production.Product INNER JOIN

                  Sales.SalesOrderDetail ON Production.Product.ProductID = Sales.SalesOrderDetail.ProductID

GROUP BY COALESCE(Production.Product.Color,'Transparente') 

ORDER BY SUM(Sales.SalesOrderDetail.OrderQty) DESC
```

\--15.- Para cada producto quiero saber cual es el mayor precio por el que hemos vendido un producto.

```sql
SELECT MIN(Sales.SalesOrderDetail.UnitPrice) AS PrecioMinimo, Production.Product.ProductID, Production.Product.Name

FROM     Production.Product INNER JOIN

                  Sales.SalesOrderDetail ON Production.Product.ProductID = Sales.SalesOrderDetail.ProductID

GROUP BY Production.Product.ProductID, Production.Product.Name
```

\--16.- Quiero ver cuanto se ha vendido de la talla (Size) 45 de color rojo. Sumar las cantidades vendidas.

```sql
SELECT SUM(Sales.SalesOrderDetail.OrderQty) AS Cantidad

FROM     Production.Product INNER JOIN

                  Sales.SalesOrderDetail ON Production.Product.ProductID = Sales.SalesOrderDetail.ProductID 

WHERE  (Production.Product.Color = 'red') and (Production.Product.Size = '45')
```

\--17.- Quiero saber cuanto se ha vendido de un producto, el problema es que mi jefe quiere que cada producto

\--salga descrito como sigue ProductID + Name + (Class - Color).

```sql
SELECT CONCAT(Production.Product.ProductID, Name, '(', Color, ')') as Nombre,  COALESCE(SUM(Sales.SalesOrderDetail.OrderQty),0) AS Cantidad

FROM     Sales.SalesOrderDetail RIGHT OUTER JOIN

                  Production.Product ON Sales.SalesOrderDetail.ProductID = Production.Product.ProductID

GROUP BY CONCAT(Production.Product.ProductID, Name, '(', Color, ')')
```
